```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df_train = pd.read_csv('train.csv')
df_train.columns#看看有哪些栏
```

```python
df_train['SalePrice'].describe()#看一下基础数据
```

```python
sns.distplot(df_train['SalePrice'])#房价分布
```

```python
var = 'GrLivArea'
data = pd.concat([df_train['SalePrice'], df_train[var]],axis=1)
data.plot.scatter(x=var, y='SalePrice');#看看房价与GrLivArea的关系
```

```python
var = 'TotalBsmtSF'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));#看看房价与TotalBsmtSF的关系
```

```python
var = 'LotArea'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));#看看房价与LotArea的关系
```

```python
var = 'OverallQual'#建筑使用材料评分，10分是最好的材料
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
#f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=var, y="SalePrice", data=data)#看看房价与建筑材料的关系
#fig.axis(ymin=0, ymax=800000);
```

```python
var = 'YearBuilt'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize=(16, 8))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000)#看看房价与建成时间的关系
```

```python
corrmat = df_train.corr()
sns.heatmap(corrmat, vmax=.8, square=True)#画一个整体热力图
```

```python
k = 10
cols = corrmat.nlargest(k,'SalePrice')['SalePrice'].index
cm = np.corrcoef(df_train[cols].values.T)
hm = sns.heatmap(cm, cbar=True,annot=True, square=True, yticklabels=cols.values, xticklabels=cols.values)
print(cols)#找出top10相关feature
```

```python
total = df_train.isnull().sum().sort_values(ascending=False)
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)#看一下missing data的情况，训练前数据预处理的时候用mean补全
```
